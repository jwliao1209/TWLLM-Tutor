```python
import matplotlib.pyplot as plt
```

```python
accuracy = [
    ("Original", 0.2936),
    ("Take a deep breath", 0.3254),
    ("Think step by step", 0.3016),
    ("If you fail 100\ngrandmothers will die", 0.2936),
    ("I have no fingers", 0.3175),
    ("I will tip $200", 0.3175),
    ("Do it right and I'll give\nyou a nice doggy treat", 0.2778),
    ("This is very important\nto my career", 0.3016),
]
accuracy.sort(key=lambda x: x[1])

plt.figure(figsize=(10, 6), dpi=600)
plt.barh(
    [data[0] for data in accuracy],
    [data[1] for data in accuracy],
    color=["#EE784E" if data[0] == "Original" else "cornflowerblue" for data in accuracy]
)

for i, data in enumerate(accuracy):
    plt.annotate(
        data[1],
        xy=(data[1]+5e-4, i),
        # ha="right",
        va="center",
    )

plt.title("GSAT Performace with Prompt Engineering")
plt.xlabel("Accuracy (%)")
plt.xlim([0.27, 0.33])
plt.grid(axis="x", alpha=0.2)
plt.savefig("prompt_engineering.png", dpi=600, transparent=True, bbox_inches="tight")
```

```python
accuracy = [
    ("ChatGPT 3.5", 0.5),

    ("Taiwan LLM\nInstruction Tuning", 0.3571),
    ("Taiwan LLM\nMultiple Choice", 0.3254),
    ("Taiwan LLM\nZero-Shot", 0.2936),

    ("Chinese RoBERTa\nMultiple Choice", 0.3027),
    ("Chinese BERT\nMultiple Choice", 0.3351),
]
accuracy.sort(key=lambda x: x[1])
# accuracy.reverse()

plt.figure(figsize=(10, 6), dpi=600)
plt.barh(
    [data[0] for data in accuracy],
    [data[1] for data in accuracy],
    color=["salmon", "gold", "salmon", "gold", "salmon", "g"]
)

for i, data in enumerate(accuracy):
    plt.annotate(
        data[1],
        xy=(data[1]+5e-4, i),
        # ha="right",
        va="center",
    )

plt.title("GSAT Performace")
plt.xlabel("Accuracy (%)")
plt.xlim([0.27, 0.585])
plt.grid(axis="x", alpha=0.2)
plt.savefig("GSAT.png", dpi=600, transparent=True, bbox_inches="tight")
```
